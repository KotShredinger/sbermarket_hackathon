```python
df_stat=pd.DataFrame(columns=['month','product_name','product_count'])
for month in products_orders.month.unique():
    df_month=pd.DataFrame(columns=['month','product_name','product_count'])
    df_month['product_name']=products_orders.loc[products_orders['month']==month]['product_name'].value_counts().head(5).index
    df_month['product_count']=products_orders.loc[products_orders['month']==month]['product_name'].value_counts().head(5).to_list()
    df_month['month']=month
    df_stat= pd.concat([df_stat, df_month], ignore_index=True)
```

```python
df_stat=pd.DataFrame(columns=['month','product_name','product_count'])
for month in [[1,2,3],[4,5,6]]:
    df_month=pd.DataFrame(columns=['month','product_name','product_count'])
    df_month['product_name']=orders.loc[(orders['month'].isin(month))]['platform'].value_counts().index
    df_month['product_count']=orders.loc[(orders['month'].isin(month))]['platform'].value_counts().to_list()
    if month == [1,2,3]:
        df_month['month']='Before'
    else:
        df_month['month']='During'
    df_stat= pd.concat([df_stat, df_month], ignore_index=True)
df_stat
```
