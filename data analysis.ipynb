```python
def ff(s):
    num=0
    for n_char,char in enumerate(s):
        if char.isdigit() and num==0:
            num=n_char
    return num
```

```python
df_stat2=df_stat1[1:11]
df_stat2['product_name']=df_stat2['product_name'].apply(lambda x: x[:ff(x)-1] if '%' in x else x)
import matplotlib.pyplot as plt

fig, ax = plt.subplots(dpi=600)

ax.bar(df_stat2['product_name'], df_stat2['corr'])
plt.ylabel('correlation')
fig.set_figwidth(20)    #  ширина Figure
fig.set_figheight(10)    #  высота Figure
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=60)
fig.autofmt_xdate()
plt.savefig('corr.jpg', bbox_inches='tight')
```
